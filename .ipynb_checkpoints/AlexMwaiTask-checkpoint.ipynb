```python
## Step 1: Import necessary libraries
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Check TensorFlow version
print(tf.__version__)
```

```python
## Step 2: Load the Malaria dataset
```

```python
import tensorflow_datasets as tfds

# Load the 'malaria' dataset
dataset, info = tfds.load('malaria', with_info=True, as_supervised=True, split='train')

# Split the 'train' dataset into training and testing datasets
train_dataset = dataset.take(int(0.7 * info.splits['train'].num_examples))
test_dataset = dataset.skip(int(0.7 * info.splits['train'].num_examples))

# Print the dataset information
print(info)
```

```python
## Step 3: Preprocess the data
```

```python
def preprocess(image, label):
    # Normalize pixel values
    image = tf.cast(image, tf.float32) / 255.0
    # Resize the images to a consistent shape
    image = tf.image.resize(image, [64, 64])
    return image, label

train_dataset = train_dataset.map(preprocess)
test_dataset = test_dataset.map(preprocess)
```

```python
## Step 4: Batch and prefetch the data
```

```python
train_dataset = train_dataset.batch(32).prefetch(tf.data.experimental.AUTOTUNE)
test_dataset = test_dataset.batch(32).prefetch(tf.data.experimental.AUTOTUNE)
```

```python
## Step 5: Define the model
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2)
])

# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
## Step 6: Train the model
```

```python
history = model.fit(train_dataset, epochs=10, validation_data=test_dataset)
```

```python
history.history
```

```python
## Step 7: Evaluate the model
```

```python
test_loss, test_acc = model.evaluate(test_dataset, verbose=2)
print('\nTest accuracy:', test_acc)
```

```python
## Step 8: Plot Accuracy of Model Accuracy & Model Loss
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_learningCurve(history, epoch):
  # Plot training & validation accuracy values
  epoch_range = range(1, epoch+1)
  plt.plot(epoch_range, history.history['accuracy'])
  plt.plot(epoch_range, history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

  # Plot training & validation loss values
  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()
```

```python
plot_learningCurve(history, 10)
```

```python
## Confussion Matrix
```

```python
predictions = model.predict(test_dataset)
predicted_labels = np.argmax(predictions, axis=1)
```

```python
true_labels = tf.concat([y for x, y in test_dataset], axis=0)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
## ROC Curve (Receiver Operating Characteristic Curve)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc



# Compute the ROC curve points
fpr, tpr, thresholds = roc_curve(true_labels, predicted_labels)

# Compute the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()
```
